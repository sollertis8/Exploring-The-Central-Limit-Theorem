```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

# Let’s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

The populations are not normal. Next, take a sample of 100 from each population and plot them.

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

```python
size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

# The squared standard deviations are divided by the sample size and summed, then we take
# the square root of the sum. 
diff_se = sum(sd ** 2 / size) ** 0.5  

#The difference between the means divided by the standard error: T-value.  
print(diff/diff_se)
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

Increasing the sample sizes from 100 to 1000, I expect to see the shape of the histograms to look more similar to that of the original populations.

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

As suspected, the shape of the samples more adequately represent their respective populations.  If we significantly decrease the sample sizes, I expect to see shapes that do not adequately represent the populations.

```python
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

As suspected, the samples do not represent the populations accurately.

```python
pop1 = np.random.binomial(10, 0.3, 10000)
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(ttest_ind(sample2, sample1, equal_var=False))

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

```python
pop1 = np.random.binomial(10, 0.4, 10000)
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(ttest_ind(sample2, sample1, equal_var=False))

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

After changing the p of the population, the t-value decreases and the p-value increases. 

```python
pop1 = np.random.poisson(5, 10000)
pop2 = np.random.poisson(7, 10000) 


plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')

plt.legend(loc='upper right')

plt.show()

print(ttest_ind(sample2, sample1, equal_var=False))

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

diff=sample2.mean() - sample1.mean()

print(diff)
```

From the samples taken from our poisson distributions, we can see that the larger the sample size, the more accurately the sample mean values represent the population values.  Conversely, the smaller the sample sizes, the less representative they are of the population.
